```python
# Group by passenger_count and calculate mean tip_amount
mean_tips_by_passenger_count = (
    df.groupby('passenger_count', as_index=False)['tip_amount']
    .mean()
)

print(mean_tips_by_passenger_count)
```

```python

# Compute mean tip by passenger count
mean_tips_by_passenger_count = (
    df.groupby('passenger_count', as_index=False)['tip_amount']
    .mean()
    .rename(columns={'tip_amount': 'mean_tip_amount'})
)

# Remove the first row if needed (e.g., passenger_count = 0)
data = mean_tips_by_passenger_count[mean_tips_by_passenger_count['passenger_count'] != 0]

# Color palette ranking
pal = sns.color_palette("Greens_d", len(data))
rank = data['mean_tip_amount'].argsort().argsort()

# Plot
plt.figure(figsize=(12, 7))
ax = sns.barplot(
    x=data['passenger_count'],
    y=data['mean_tip_amount'],
    palette=np.array(pal[::-1])[rank]
)

# Global mean line
ax.axhline(df['tip_amount'].mean(), ls='--', color='red', label='Global Mean')

# Labels and title
ax.set_xlabel('Passenger Count')
ax.set_ylabel('Mean Tip Amount')
plt.title('Mean Tip Amount by Passenger Count', fontsize=16)
ax.legend()
plt.tight_layout()
plt.show()
```

```python
# Create a month column
df['month'] = df['tpep_pickup_datetime'].dt.month_name()
# Create a day column
df['day'] = df['tpep_pickup_datetime'].dt.day_name()
```

```python
# Get total number of rides for each month
monthly_rides = df['month'].value_counts()
monthly_rides
```

```python
# Reorder the monthly ride list so months go in order
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July',
         'August', 'September', 'October', 'November', 'December']

monthly_rides = monthly_rides.reindex(index=month_order)
monthly_rides
```

```python
# Show the index
monthly_rides.index
```

```python


# Create a bar plot of total rides per month
plt.figure(figsize=(12,7))
ax = sns.barplot(x=monthly_rides.index, y=monthly_rides)
ax.set_xticklabels(month_order)
plt.title('Ride count by month', fontsize=15);
```

```python
# Repeat the above process, this time for rides by day
daily_rides = df['day'].value_counts()
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
daily_rides = daily_rides.reindex(index=day_order)
daily_rides
```

```python
# Create bar plot for ride count by day
plt.figure(figsize=(12,7))
ax = sns.barplot(x=daily_rides.index, y=daily_rides)
ax.set_xticklabels(day_order)
ax.set_ylabel('Count')
plt.title('Ride count by day', fontsize=16);
```

```python
# Get number of unique drop-off location IDs
df['DOLocationID'].nunique()
```

```python
# Calculate mean trip distance for each drop-off location
distance_by_dropoff = (
    df.groupby('DOLocationID', as_index=True)['trip_distance']
    .mean()
    .sort_values(ascending=False)
    .reset_index()
    .rename(columns={'trip_distance': 'mean_trip_distance'})
)

# Display the result
print(distance_by_dropoff)
```

```python
# Create a bar plot of mean trip distances by drop-off location (ascending order)
plt.figure(figsize=(14, 6))
ax = sns.barplot(
    x='DOLocationID',
    y='mean_trip_distance',
    data=distance_by_dropoff.sort_values(by='mean_trip_distance', ascending=True),
    palette='coolwarm'
)

# Remove x-axis ticks and labels for clarity (optional)
ax.set_xticklabels([])
ax.set_xticks([])

# Title and axis labels
plt.title('Mean Trip Distance by Drop-off Location', fontsize=16)
plt.xlabel('Drop-off Location ID')
plt.ylabel('Mean Trip Distance')
plt.tight_layout()
plt.show()
```

```python
# 1. Generate random points from a normal distribution
test = np.round(np.random.normal(10, 5, (3000, 2)), 1)
midway = int(len(test)/2)
start = test[:midway]
end = test[midway:]

# 2. Calculate Euclidean distances
distances = np.sqrt(((start - end) ** 2).sum(axis=1))

# 3. Create DataFrame and group by drop-off location
test_df = pd.DataFrame({
    'start': [tuple(x) for x in start],
    'end': [tuple(x) for x in end],
    'distance': distances
})

# 4. Compute mean distance per end location
data = (
    test_df[['end', 'distance']]
    .groupby('end')
    .mean()
    .sort_values(by='distance')
    .reset_index()
)

# Convert tuple to string for plotting
data['end_str'] = data['end'].astype(str)

# 5. Plot
plt.figure(figsize=(14, 6))
ax = sns.barplot(x='end_str', y='distance', data=data, palette='viridis')

# Clean up x-axis for clarity
ax.set_xticks([])
ax.set_xticklabels([])
ax.set_xlabel('Drop-off Location (hidden for clarity)')
ax.set_ylabel('Mean Distance')
ax.set_title('Mean Distance Between Randomly Generated Drop-off Points')

plt.tight_layout()
plt.show()
```
